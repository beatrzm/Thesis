```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from scipy import stats
import os
```

```python
numbers_controls = range(19, 52)

# Dictionary to store DataFrames
dataframes_mid = {}

for i in numbers_controls:
    sub_control = f"sub-control{i:03d}"
    filename = f"featqueryreport_{sub_control}_ses-midcycle.txt"
    filepath = os.path.join("C:\\Users\\beama\\OneDrive\\Documents\\uni\\4 ano\\Tese\\plots\\all_Lamyg_reports", filename)
    if os.path.exists(filepath):
        dataframes_mid[f"control{i}_ses-midcycle"] = pd.read_csv(filepath, sep=" ", header=None)
        #print(f"Successfully read {filename}. Shape: {dataframes_mid[f'control{i}_{phase}'].shape}")
    else:
        #print(f"File {filename} not found. Skipping...")
        pass

print(len(dataframes_mid))

# Dictionary to store DataFrames
dataframes_premens = {}

for i in numbers_controls:
    sub_control = f"sub-control{i:03d}"
    filename = f"featqueryreport_{sub_control}_ses-premenstrual.txt"
    filepath = os.path.join("C:\\Users\\beama\\OneDrive\\Documents\\uni\\4 ano\\Tese\\plots\\all_Lamyg_reports", filename)
    if os.path.exists(filepath):
        dataframes_premens[f"control{i}_ses-premenstrual"] = pd.read_csv(filepath, sep=" ", header=None)
        #print(f"Successfully read {filename}. Shape: {dataframes_premens[f'control{i}_{phase}'].shape}")
    else:
        #print(f"File {filename} not found. Skipping...")
        pass

print(len(dataframes_premens))
```

```python
numbers_patients = range(1, 53)

dataframes_int = {}

for i in numbers_patients:
    sub_patient = f"sub-patient{i:03d}"
    filename = f"featqueryreport_{sub_patient}_ses-interictal.txt"
    filepath = os.path.join("C:\\Users\\beama\\OneDrive\\Documents\\uni\\4 ano\\Tese\\plots\\all_Lamyg_reports", filename)
    if os.path.exists(filepath):
        dataframes_int[f"patient{i}_ses-interictal"] = pd.read_csv(filepath, sep=" ", header=None)
        #print(f"Successfully read {filename}. Shape: {dataframes_int[f'patient{i}_{phase}'].shape}")
    else:
        #print(f"File {filename} not found. Skipping...")
        pass

print(len(dataframes_int))

# Dictionary to store DataFrames
dataframes_preictal = {}

for i in numbers_patients:
    sub_patient = f"sub-patient{i:03d}"
    filename = f"featqueryreport_{sub_patient}_ses-preictal.txt"
    filepath = os.path.join("C:\\Users\\beama\\OneDrive\\Documents\\uni\\4 ano\\Tese\\plots\\all_Lamyg_reports", filename)
    if os.path.exists(filepath):
        dataframes_preictal[f"patient{i}_ses-preictal"] = pd.read_csv(filepath, sep=" ", header=None)
        #print(f"Successfully read {filename}. Shape: {dataframes_preictal[f'patient{i}_{phase}'].shape}")
    else:
        #print(f"File {filename} not found. Skipping...")
        pass

print(len(dataframes_preictal))

# Dictionary to store DataFrames
dataframes_ictal = {}

for i in numbers_patients:
    sub_patient = f"sub-patient{i:03d}"
    filename = f"featqueryreport_{sub_patient}_ses-ictal.txt"
    filepath = os.path.join("C:\\Users\\beama\\OneDrive\\Documents\\uni\\4 ano\\Tese\\plots\\all_Lamyg_reports", filename)
    if os.path.exists(filepath):
        dataframes_ictal[f"patient{i}_ses-ictal"] = pd.read_csv(filepath, sep=" ", header=None)
        #print(f"Successfully read {filename}. Shape: {dataframes_ictal[f'patient{i}_{phase}'].shape}")
    else:
        #print(f"File {filename} not found. Skipping...")
        pass

print(len(dataframes_ictal))

# Dictionary to store DataFrames
dataframes_postictal = {}

for i in numbers_patients:
    sub_patient = f"sub-patient{i:03d}"
    filename = f"featqueryreport_{sub_patient}_ses-postictal.txt"
    filepath = os.path.join("C:\\Users\\beama\\OneDrive\\Documents\\uni\\4 ano\\Tese\\plots\\all_Lamyg_reports", filename)
    if os.path.exists(filepath):
        dataframes_postictal[f"patient{i}_ses-postictal"] = pd.read_csv(filepath, sep=" ", header=None)
        #print(f"Successfully read {filename}. Shape: {dataframes_postictal[f'patient{i}_{phase}'].shape}")
    else:
        #print(f"File {filename} not found. Skipping...")
        pass

print(len(dataframes_postictal))
```

```python
# for each dataframe in each dictionary get the value in .iloc[0,5] and store it in a list

midcycle = []
for key in dataframes_mid:
    midcycle.append(dataframes_mid[key].iloc[0,5])

premenstrual = []
for key in dataframes_premens:
    premenstrual.append(dataframes_premens[key].iloc[0,5])

interictal = []
for key in dataframes_int:
    interictal.append(dataframes_int[key].iloc[0,5])

preictal = []
for key in dataframes_preictal:
    preictal.append(dataframes_preictal[key].iloc[0,5])

ictal = []
for key in dataframes_ictal:
    ictal.append(dataframes_ictal[key].iloc[0,5])

postictal = []
for key in dataframes_postictal:
    postictal.append(dataframes_postictal[key].iloc[0,5])

print(midcycle)
```

```python
# test for significance

midvsinter8 = stats.ttest_ind(midcycle, interictal)
premensvspreict8 = stats.ttest_ind(premenstrual, preictal)
premensvsict8 = stats.ttest_ind(premenstrual, ictal)
premensvspostict8 = stats.ttest_ind(premenstrual, postictal)
midvspremens8 = stats.ttest_rel(midcycle, premenstrual)

print("Midcycle vs Interictal unpaired t-test for COPE8: ",midvsinter8)
print("Premenstrual vs Preictal unpaired t-test for COPE8: ",premensvspreict8)
print("Premenstrual vs Ictal unpaired t-test for COPE8: ",premensvsict8)
print("Premenstrual vs Postictal unpaired t-test for COPE8: ",premensvspostict8)
print("Midcycle vs Premenstrual paired t-test for COPE8: ",midvspremens8)

```

```python
import matplotlib.pyplot as plt

# Define the data
data = [midcycle, premenstrual, interictal, preictal, ictal, postictal]

# Define the positions for the boxplots
positions = [1, 2, 3, 4, 5, 6]  # Positions for each phase

# Create the boxplot with equal width
boxprops_green = dict(color="lightgreen", facecolor="lightgreen", linewidth=2)  # Define properties for light green color
boxprops_red = dict(color="lightcoral", facecolor="lightcoral", linewidth=2)  # Define properties for light red color
medianprops_black = dict(color="black", linewidth=2)  # Define properties for median line

plt.boxplot(data[:-2], positions=positions[:-2], patch_artist=True, boxprops=boxprops_green, medianprops=medianprops_black,widths=0.5)  # Plot first two boxplots
plt.boxplot(data[-4:], positions=positions[-4:], patch_artist=True, boxprops=boxprops_red, medianprops=medianprops_black, widths=0.5)  # Plot last four boxplots

# Set x-axis labels
phase_names = ['Midcycle', 'Premenstrual', 'Interictal', 'Preictal', 'Ictal', 'Postictal']
plt.xticks(positions, phase_names, fontsize=8)

# sig bars
max_value = max(max(midcycle), max(premenstrual), max(interictal), max(preictal), max(ictal), max(postictal))
print(max_value)
offset = 0.02  # Offset to position the bars slightly above the maximum value
if midvsinter8.pvalue < 0.05:
    plt.hlines(max_value + offset, positions[0], positions[2], color='black', linewidth=2)
    plt.text((positions[0] + positions[2]) * 0.5, max_value + 1 * offset, '*', ha='center', va='bottom', fontsize=16)
if premensvspreict8.pvalue < 0.05:
    plt.hlines(max_value + offset, positions[1], positions[3], color='black', linewidth=2)
    plt.text((positions[1] + positions[3]) * 0.5, max_value + 2 * offset, '*', ha='center', va='bottom', fontsize=16)
if premensvsict8.pvalue < 0.05:
    plt.hlines(max_value + offset, positions[1], positions[4], color='black', linewidth=2)
    plt.text((positions[1] + positions[4]) * 0.5, max_value + 2 * offset, '*', ha='center', va='bottom', fontsize=16)
if premensvspostict8.pvalue < 0.05:
    plt.hlines(max_value + offset, positions[1], positions[5], color='black', linewidth=2)
    plt.text((positions[1] + positions[5]) * 0.5, max_value + 2 * offset, '*', ha='center', va='bottom', fontsize=16)
if midvspremens8.pvalue < 0.05:
    plt.hlines(max_value + offset, positions[0], positions[1], color='black', linewidth=2)
    plt.text((positions[0] + positions[1]) * 0.5, max_value +0.8 * offset, '*', ha='center', va='bottom', fontsize=16)

plt.ylabel('Mean % signal change')
plt.title('Mean % signal change for COPE8 (H>N) in the left amygdala (stats map)')
plt.show()


```
