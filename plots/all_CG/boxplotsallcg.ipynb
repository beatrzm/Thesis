```python
def extract_values(dataframes, iloc_position):
    extracted_values = []
    for key in dataframes:
        extracted_values.append(dataframes[key].iloc[iloc_position[0], iloc_position[1]])
    return extracted_values

# extract values for each phase in each roi

midcycleACC = extract_values(dataframes_midACC, (0, 5))
premenstrualACC = extract_values(dataframes_premensACC, (0, 5))
interictalACC = extract_values(dataframes_intACC, (0, 5))
preictalACC = extract_values(dataframes_preictalACC, (0, 5))
ictalACC = extract_values(dataframes_ictalACC, (0, 5))
postictalACC = extract_values(dataframes_postictalACC, (0, 5))

midcycleMCC = extract_values(dataframes_midMCC, (0, 5))
premenstrualMCC = extract_values(dataframes_premensMCC, (0, 5))
interictalMCC = extract_values(dataframes_intMCC, (0, 5))
preictalMCC = extract_values(dataframes_preictalMCC, (0, 5))
ictalMCC = extract_values(dataframes_ictalMCC, (0, 5))
postictalMCC = extract_values(dataframes_postictalMCC, (0, 5))

midcyclePCC = extract_values(dataframes_midPCC, (0, 5))
premenstrualPCC = extract_values(dataframes_premensPCC, (0, 5))
interictalPCC = extract_values(dataframes_intPCC, (0, 5))
preictalPCC = extract_values(dataframes_preictalPCC, (0, 5))
ictalPCC = extract_values(dataframes_ictalPCC, (0, 5))
postictalPCC = extract_values(dataframes_postictalPCC, (0, 5))

```

```python
formatdata = pd.DataFrame({
    'Phase': ['Midcycle'] * len(midcycleACC) + ['Premenstrual'] * len(premenstrualACC) + ['Interictal'] * len(interictalACC) + \
             ['Midcycle'] * len(midcycleMCC) + ['Premenstrual'] * len(premenstrualMCC) + ['Interictal'] * len(interictalMCC) + \
             ['Midcycle'] * len(midcyclePCC) + ['Premenstrual'] * len(premenstrualPCC) + ['Interictal'] * len(interictalPCC) + \
             ['Preictal'] * len(preictalACC) + ['Ictal'] * len(ictalACC) + ['Postictal'] * len(postictalACC) + \
             ['Preictal'] * len(preictalMCC) + ['Ictal'] * len(ictalMCC) + ['Postictal'] * len(postictalMCC) + \
             ['Preictal'] * len(preictalPCC) + ['Ictal'] * len(ictalPCC) + ['Postictal'] * len(postictalPCC),
    'ROI': ['ACC'] * len(midcycleACC) + ['ACC'] * len(premenstrualACC) + ['ACC'] * len(interictalACC) + \
           ['MCC'] * len(midcycleMCC) + ['MCC'] * len(premenstrualMCC) + ['MCC'] * len(interictalMCC) + \
           ['PCC'] * len(midcyclePCC) + ['PCC'] * len(premenstrualPCC) + ['PCC'] * len(interictalPCC) + \
           ['ACC'] * len(preictalACC) + ['ACC'] * len(ictalACC) + ['ACC'] * len(postictalACC) + \
           ['MCC'] * len(preictalMCC) + ['MCC'] * len(ictalMCC) + ['MCC'] * len(postictalMCC) + \
           ['PCC'] * len(preictalPCC) + ['PCC'] * len(ictalPCC) + ['PCC'] * len(postictalPCC),
    'Mean % Signal Change': midcycleACC + premenstrualACC + interictalACC +
                            midcycleMCC + premenstrualMCC + interictalMCC +
                            midcyclePCC + premenstrualPCC + interictalPCC +
                            preictalACC + ictalACC + postictalACC +
                            preictalMCC + ictalMCC + postictalMCC +
                            preictalPCC + ictalPCC + postictalPCC
})

print(formatdata.head(10))
```
