```python
def extract_values(dataframes, iloc_position):
    extracted_values = []
    for key in dataframes:
        extracted_values.append(dataframes[key].iloc[iloc_position[0], iloc_position[1]])
    return extracted_values

# extract values for each phase in each roi

midcycleOFG = extract_values(dataframes_midOFG, (0, 5))
premenstrualOFG = extract_values(dataframes_premensOFG, (0, 5))
interictalOFG = extract_values(dataframes_intOFG, (0, 5))
preictalOFG = extract_values(dataframes_preictalOFG, (0, 5))
ictalOFG = extract_values(dataframes_ictalOFG, (0, 5))
postictalOFG = extract_values(dataframes_postictalOFG, (0, 5))

midcycleLG = extract_values(dataframes_midLG, (0, 5))
premenstrualLG = extract_values(dataframes_premensLG, (0, 5))
interictalLG = extract_values(dataframes_intLG, (0, 5))
preictalLG = extract_values(dataframes_preictalLG, (0, 5))
ictalLG = extract_values(dataframes_ictalLG, (0, 5))
postictalLG = extract_values(dataframes_postictalLG, (0, 5))

midcyclePCC = extract_values(dataframes_midPCC, (0, 5))
premenstrualPCC = extract_values(dataframes_premensPCC, (0, 5))
interictalPCC = extract_values(dataframes_intPCC, (0, 5))
preictalPCC = extract_values(dataframes_preictalPCC, (0, 5))
ictalPCC = extract_values(dataframes_ictalPCC, (0, 5))
postictalPCC = extract_values(dataframes_postictalPCC, (0, 5))

```

```python
formatdata = pd.DataFrame({
    'Phase': ['Midcycle'] * len(midcycleOFG) + ['Premenstrual'] * len(premenstrualOFG) + ['Interictal'] * len(interictalOFG) + \
             ['Midcycle'] * len(midcycleLG) + ['Premenstrual'] * len(premenstrualLG) + ['Interictal'] * len(interictalLG) + \
             ['Midcycle'] * len(midcyclePCC) + ['Premenstrual'] * len(premenstrualPCC) + ['Interictal'] * len(interictalPCC) + \
             ['Preictal'] * len(preictalOFG) + ['Ictal'] * len(ictalOFG) + ['Postictal'] * len(postictalOFG) + \
             ['Preictal'] * len(preictalLG) + ['Ictal'] * len(ictalLG) + ['Postictal'] * len(postictalLG) + \
             ['Preictal'] * len(preictalPCC) + ['Ictal'] * len(ictalPCC) + ['Postictal'] * len(postictalPCC),
    'ROI': ['OFG'] * len(midcycleOFG) + ['OFG'] * len(premenstrualOFG) + ['OFG'] * len(interictalOFG) + \
           ['LG'] * len(midcycleLG) + ['LG'] * len(premenstrualLG) + ['LG'] * len(interictalLG) + \
           ['PCC'] * len(midcyclePCC) + ['PCC'] * len(premenstrualPCC) + ['PCC'] * len(interictalPCC) + \
           ['OFG'] * len(preictalOFG) + ['OFG'] * len(ictalOFG) + ['OFG'] * len(postictalOFG) + \
           ['LG'] * len(preictalLG) + ['LG'] * len(ictalLG) + ['LG'] * len(postictalLG) + \
           ['PCC'] * len(preictalPCC) + ['PCC'] * len(ictalPCC) + ['PCC'] * len(postictalPCC),
    'Mean % Signal Change': midcycleOFG + premenstrualOFG + interictalOFG +
                            midcycleLG + premenstrualLG + interictalLG +
                            midcyclePCC + premenstrualPCC + interictalPCC +
                            preictalOFG + ictalOFG + postictalOFG +
                            preictalLG + ictalLG + postictalLG +
                            preictalPCC + ictalPCC + postictalPCC
})

print(formatdata.head(10))
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=formatdata, x='Phase', y='Mean % Signal Change', hue='ROI')

# Add labels and title
plt.xlabel('Phase')
plt.ylabel('Mean % Signal Change')
plt.title('Mean % Signal Change Across Phases and ROIs for COPE4 (N,H,A)')
#plt.ylim(-2,2)

# Show legend
plt.legend(title='ROI')

# Show plot
plt.show()
```
